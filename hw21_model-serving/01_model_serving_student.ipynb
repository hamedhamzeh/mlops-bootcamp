## Part 1 — Model Reproducibility Check

Before serving a model, you must verify it produces exactly the same output as it did during training.

This is called a **reproducibility check** and it catches silent bugs like:
- preprocessing mismatch between training and serving
- wrong model version loaded
- feature column order changed

### 1.1 Connect to MLflow and load your best model

````python
MLFLOW_TRACKING_URI = os.getenv('MLFLOW_TRACKING_URI', 'http://185.50.38.163:33014')
MLFLOW_USERNAME = os.getenv('MLFLOW_TRACKING_USERNAME', '') or input('MLflow username: ').strip()
MLFLOW_PASSWORD = os.getenv('MLFLOW_TRACKING_PASSWORD', '') or input('MLflow password: ').strip()

os.environ['MLFLOW_TRACKING_USERNAME'] = MLFLOW_USERNAME
os.environ['MLFLOW_TRACKING_PASSWORD'] = MLFLOW_PASSWORD

mlflow.set_tracking_uri(MLFLOW_TRACKING_URI)
client = MlflowClient()

experiment = client.get_experiment_by_name(EXPERIMENT_NAME)
if experiment is None:
    raise ValueError(f'Experiment not found: {EXPERIMENT_NAME}. Complete HW02 first.')

runs = mlflow.search_runs(
    experiment_ids=[experiment.experiment_id],
    filter_string="tags.leakage_status = 'clean' and tags.selected_for_serving = 'true'",
    order_by=['metrics.f1 DESC'],
)

if runs.empty:
    raise ValueError(
        'No run tagged selected_for_serving=true found. '
        'Go to MLflow UI, find your best clean run, and add the tag.'
    )

BEST_RUN_ID = runs.iloc[0]['run_id']
MODEL_URI = f'runs:/{BEST_RUN_ID}/model'

print('Best run ID  :', BEST_RUN_ID)
print('Model URI    :', MODEL_URI)
print('Run name     :', runs.iloc[0].get('tags.mlflow.runName'))
print('F1 score     :', runs.iloc[0].get('metrics.f1'))
````

### 2.6 Batch vs single benchmark

**TODO 2.6**

Take 100 rows from your feature dataset.

Measure:
1. Time to call `POST /predict` 100 times in a loop (single)
2. Time to call `POST /predict/batch` once with all 100 rows (batch)

Print a comparison table with total time and time per prediction for each approach.

Then answer: why is batch faster? Write your answer as a comment in the cell.

````python
import math
import time
import requests
import pandas as pd

def clean_for_json(row: dict) -> dict:
    return {
        k: None if isinstance(v, float) and math.isnan(v) else v
        for k, v in row.items()
    }

BENCHMARK_SIZE = 100
benchmark_rows = [
    clean_for_json(row)
    for row in df[FEATURE_COLS].head(BENCHMARK_SIZE).to_dict(orient='records')
]

# Single prediction benchmark
single_start = time.perf_counter()

single_responses = []

for row in benchmark_rows:
    response = requests.post(
        f"{BASE_URL}/predict",
        json=row,
        timeout=30,
    )

    assert response.status_code == 200, (
        f"Single prediction failed. "
        f"Status: {response.status_code}. "
        f"Response: {response.text}"
    )

    single_responses.append(response.json())

single_total = time.perf_counter() - single_start
single_per_prediction_ms = (single_total / BENCHMARK_SIZE) * 1000


# Batch prediction benchmark
batch_start = time.perf_counter()

batch_response = requests.post(
    f"{BASE_URL}/predict/batch",
    json=benchmark_rows,
    timeout=60,
)

assert batch_response.status_code == 200, (
    f"Batch prediction failed. "
    f"Status: {batch_response.status_code}. "
    f"Response: {batch_response.text}"
)

batch_results = batch_response.json()

batch_total = time.perf_counter() - batch_start
batch_per_prediction_ms = (batch_total / BENCHMARK_SIZE) * 1000


# Validation
assert len(single_responses) == BENCHMARK_SIZE, "Single response count mismatch."
assert len(batch_results) == BENCHMARK_SIZE, "Batch response count mismatch."

single_predictions = [item["prediction"] for item in single_responses]
batch_predictions = [item["prediction"] for item in batch_results]

assert single_predictions == batch_predictions, (
    "Single and batch predictions are not identical."
)


# 4. Results
speedup = single_total / batch_total if batch_total > 0 else float("inf")

comparison_df = pd.DataFrame(
    [
        {
            "Method": "single loop",
            "Total (s)": single_total,
            "Per prediction (ms)": single_per_prediction_ms,
        },
        {
            "Method": "batch",
            "Total (s)": batch_total,
            "Per prediction (ms)": batch_per_prediction_ms,
        },
    ]
)

print("Method        | Total (s) | Per prediction (ms)")
print("--------------|-----------|--------------------")

for _, row in comparison_df.iterrows():
    print(
        f"{row['Method']:<13} | "
        f"{row['Total (s)']:>9.4f} | "
        f"{row['Per prediction (ms)']:>18.2f}"
    )

print(f"\nSpeedup: {speedup:.2f}x")

````

Method        | Total (s) | Per prediction (ms)
--------------|-----------|--------------------
single loop   |  245.6273 |            2456.27
batch         |    2.5735 |              25.74

Speedup: 95.44x

Batch is faster because it sends one HTTP request instead of 100 separate requests.
It also builds one DataFrame and calls model.predict/model.predict_proba once for all rows.
The single loop repeats request handling, validation, DataFrame creation, and model inference 100 times.

**TODO 4.3 — Answer here:**

**Q1.** We set `replicas: 2` instead of 1. What happens to traffic if one Pod crashes while replicas is 1 vs 2?

A: With one replica, the service may become unavailable if the only Pod crashes. With two replicas, Kubernetes can still send traffic to the remaining healthy Pod while it replaces the failed one.

**Q2.** The `readinessProbe` has `initialDelaySeconds: 15`. Why do we need a delay specifically for this service?

A: Because the service needs time to download and load the model from MLflow at startup. The delay prevents Kubernetes from sending traffic before the model is ready.

**Q3.** Why do we store credentials in a Kubernetes Secret instead of writing them directly in `deployment.yaml`?

A: In order to reduces the risk of exposing credentials in Git or shared YAML files.

**Q4.** What is the difference between `ClusterIP` and `LoadBalancer` service types? When would you use each?

A: ClusterIP exposes the service only inside the Kubernetes cluster, which is useful for internal services. LoadBalancer exposes the service externally, so Its better to be used when users or systems outside the cluster need to access the API.
